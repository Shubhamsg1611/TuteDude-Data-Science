# Titanic Ship Passengers Survival Prediction



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
titanic = pd.read_csv(r"C:\Users\shubh\Downloads\Titanic-Dataset.csv")
titanic.head(3)
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
titanic.isnull().sum()
```

```python
titanic['Age'].fillna(titanic['Age'].median(), inplace=True)
titanic.head()
```

```python
titanic['Embarked'].fillna(titanic['Embarked'].mode()[0], inplace=True)
titanic.head()
```

```python
titanic.drop(['PassengerId','Cabin','Name','Ticket'], axis=1, inplace=True)
titanic.head()
```

```python
titanic['Sex'] = titanic['Sex'].map({'male':1, 'female':0})
titanic['Embarked'] = titanic['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})
titanic.head()
```

```python
titanic['FamSize'] = titanic['SibSp']+titanic['Parch']
titanic.head()
```

```python
titanic.drop(['SibSp','Parch'], axis=1, inplace=True)
titanic.head()
```

```python
current_columns = titanic.columns.tolist()

new_column_order = ['Survived', 'Pclass']
new_column_order.append('FamSize')

remaining_columns = ['Sex', 'Age', 'Fare', 'Embarked']
new_column_order.extend(remaining_columns)

for col in current_columns:
    if col not in new_column_order:
        new_column_order.append(col)
titanic = titanic[new_column_order]
titanic
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(6, 4))
sns.countplot(data=titanic, x="Pclass", hue="Survived")
plt.title("Survival Rate by Passenger Class")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=titanic, x="Sex", hue="Survived")
plt.title("Survival Rate by Gender")
plt.xticks(ticks=[0, 1], labels=["Female", "Male"])  # Change x-axis labels
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(data=titanic, x="Age", hue="Survived", multiple="stack", kde=True)
plt.title("Age Distribution by Survival Status")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=titanic, x="Pclass", y="Fare")
plt.title("Fare Distribution by Passenger Class")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=titanic, x="Embarked")
plt.title("Passenger Count by Embarkation Port")
plt.xticks(ticks=[0, 1, 2], labels=["Cherbourg", "Queenstown", "Southampton"])  # Map numeric to actual names
plt.show()
```

```python
x = titanic.drop('Survived', axis=1)
x
```

```python
y = titanic['Survived']
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
report = classification_report(y_test, y_pred)
```

```python
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
print(f"Accuracy: {accuracy:.4f}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(report)
```

```python
def predict_survival(pclass, famsize, sex, age, fare, embarked, model):
    if pclass not in [1, 2, 3] or sex not in [0, 1] or embarked not in [0, 1, 2]:
        return "Invalid input: Check pclass (1-3), sex (0-Female, 1-Male), and embarked (0-C, 1-Q, 2-S)."

    input_data = pd.DataFrame([[pclass, famsize, sex, age, fare, embarked]], 
                              columns=['Pclass', 'FamSize', 'Sex', 'Age', 'Fare', 'Embarked'])

    prediction = model.predict(input_data)[0]
    return "Survived" if prediction == 1 else "Not Survived"

# Get user input
try:
    pclass = int(input("Enter Passenger Class (1, 2, 3): "))
    famsize = int(input("Enter Family Size (SibSp + Parch): "))
    sex = int(input("Enter Sex (0 for Female, 1 for Male): "))
    age = float(input("Enter Age: "))
    embarked = int(input("Enter Embarkation (0-Cherbourg, 1-Queenstown, 2-Southampton): "))

    # Predict survival
    result = predict_survival(pclass, famsize, sex, age, fare, embarked, model)
    print(f"\nPrediction Result: {result}")

except ValueError:
    print("\nInvalid input! Please enter numeric values where required.")
```
