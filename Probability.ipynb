```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# Task 1a: Simulating a coin toss

def coin_toss_simulation(n=10000):
    results = [random.choice(['H', 'T']) for _ in range(n)]
    heads_prob = results.count('H') / n
    tails_prob = results.count('T') / n
    return heads_prob, tails_prob

# Task 1b: Rolling two dice and computing probability of sum 7

def dice_roll_simulation(n=10000):
    count = sum(1 for _ in range(n) if random.randint(1, 6) + random.randint(1, 6) == 7)
    return count / n

# Result:
print("Coin Toss Probabilities (Heads, Tails):", coin_toss_simulation())
print("Probability of rolling sum 7:", dice_roll_simulation())
```

```python
# Task 2: Probability of getting at least one '6' in 10 rolls

def at_least_one_six(n=10000):
    count = sum(1 for _ in range(n) if 6 in [random.randint(1, 6) for _ in range(10)])
    return count / n

print("Probability of getting at least one 6 in 10 rolls:", at_least_one_six())
```

```python
# Task 3: Conditional probability of drawing red given previous was blue

def conditional_probability_simulation(n=1000):
    balls = ['R'] * 5 + ['G'] * 7 + ['B'] * 8
    previous_blue = 0
    red_given_blue = 0
    
    prev_ball = random.choice(balls)
    for _ in range(n):
        current_ball = random.choice(balls)
        if prev_ball == 'B':
            previous_blue += 1
            if current_ball == 'R':
                red_given_blue += 1
        prev_ball = current_ball
    
    return red_given_blue / previous_blue if previous_blue > 0 else 0

print("Conditional Probability P(Red | Blue):", conditional_probability_simulation())
```

```python
# Task 4: Discrete random variable statistics

def discrete_random_variable_statistics(n=1000):
    values = [1, 2, 3]
    probabilities = [0.25, 0.35, 0.4]
    sample = np.random.choice(values, size=n, p=probabilities)
    mean = np.mean(sample)
    variance = np.var(sample)
    std_dev = np.std(sample)
    return mean, variance, std_dev

print("Discrete RV Statistics (Mean, Variance, Std Dev):", discrete_random_variable_statistics())
```

```python
# Task 5: Exponential distribution visualization

def exponential_distribution_simulation(n=2000, mean=5):
    samples = np.random.exponential(scale=mean, size=n)
    plt.hist(samples, bins=30, density=True, alpha=0.6, color='g', label='Histogram')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.title('Exponential Distribution')
    plt.legend()
    plt.show()

exponential_distribution_simulation()
```

```python
# Task 6: Central Limit Theorem Simulation

def central_limit_theorem_simulation(n_samples=1000, sample_size=30):
    population = np.random.uniform(0, 1, 10000)
    sample_means = [np.mean(np.random.choice(population, sample_size)) for _ in range(n_samples)]
    
    plt.hist(population, bins=30, alpha=0.5, label='Original Uniform Distribution', color='blue')
    plt.hist(sample_means, bins=30, alpha=0.7, label='Sample Mean Distribution', color='red')
    plt.legend()
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('Central Limit Theorem Simulation')
    plt.show()

central_limit_theorem_simulation()
```
